```python
import math

class Shape:
    def __init__(self, side_length):
        self.side_length = side_length

    def display(self, shape_name, area, perimeter=None, angles_sum=None):
        print(f"{shape_name}:")
        print(f"Area: {area}")
        if perimeter is not None:
            print(f"Perimeter: {perimeter}")
        if angles_sum is not None:
            print(f"Sum of Angles: {angles_sum}")

class Hexagon(Shape):
    def __init__(self, side_length):
        super().__init__(side_length)

    def calcArea(self):
        return 1.5 * 1.732 * self.side_length ** 2

    def calcPeri(self):
        return 6 * self.side_length

    def calcAngleSum(self):
        return 6 * 120

    def display(self):
        super().display("Hexagon", self.calcArea(), self.calcPeri(), self.calcAngleSum())

class Square(Shape):
    def __init__(self, side_length):
        super().__init__(side_length)

    def calcAreaSquare(self):
        return self.side_length ** 2

    def calcPeriSquare(self):
        return 4 * self.side_length

    def display(self):
        super().display("Square", self.calcAreaSquare(), self.calcPeriSquare())

def main():
    cnic_last_digit = int(input("Enter the last digit of your CNIC: "))

    while True:
        choice = input("Enter 1 for Hexagon, 2 for Square, or any other key to exit: ")

        if choice == '1':
            hexagon = Hexagon(cnic_last_digit)
            hexagon.display()
        elif choice == '2':
            square = Square(cnic_last_digit + 1)
            square.display()
        else:
            print("Program terminated.")
            break

if __name__ == "__main__":
    main()
```
